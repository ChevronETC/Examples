```julia
using PyPlot, JetPack, Jets, JetPackDSP
```

```julia
v = read!("../marmousi_20m_nz151_nx461.bin", Array{Float32}(undef, 151,461))
dz,dx = 0.02,0.02
nz,nx = size(v)
@show dz,dx
@show nz,nx;
```

```julia
figure(figsize=(10,5))
imshow(v,aspect="auto");
colorbar();
title("marmousi (km/s)");
```

## Smooth model using convolutional smoothing

```julia
?JopConvolve
```

```julia
a = 10
b = a+1
P = JopPad(JetSpace(Float64,nz,nx), -b:nz+b, -b:nx+b, extend=true)
S = JopConvolve(range(P), smoother=:rect, n=(a,a))
R = JopPad(JetSpace(Float64,nz,nx), -b:nz+b, -b:nx+b, extend=false)
vsm = R'∘S∘P*(1 ./ v);
vsm = 1 ./ vsm;
```

```julia
figure(figsize=(10,5))
imshow(vsm,aspect="auto");
colorbar();
title("marmousi (km/s)");
```

```julia
write("../marmousi_smoothed_20m_nz151_nx461.bin",vsm);
```
